```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#import the datasets
df1 = pd.read_csv("2006.csv.bz2")
df2 = pd.read_csv("2007.csv.bz2")
```

df1.head()

```python
#merging the datasets
df = pd.concat([df1,df2])
df
```

```python
df=df.reset_index(drop=True)
```

```python
#removing unneccesary columns 
df = df.drop(['Diverted','CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay','CancellationCode','Cancelled'], axis='columns')
```

```python
#it was observed that dep time and arr time had some erronous variables
df = df.loc[(df["DepTime"] <= 2359) & (df["ArrTime"] <= 2359)] #get rid of erronous times
```

```python
df=df.dropna()
df
```

```python
df = df.drop_duplicates()
```

```python
df.shape #35 duplicated rows removed
```

```python
#exporting cleaned data to a new csv
df.to_csv("cleaned_data.csv")
```
